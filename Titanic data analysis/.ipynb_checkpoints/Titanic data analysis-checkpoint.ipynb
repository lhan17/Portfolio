# Data analysis on Titanic dataset

## Titanic: data preprocessing and imputation

Data is downloaded from [Titanic dataset](https://www.kaggle.com/c/titanic) from Kaggle.

The dataset consists of personal information of all the passengers on board the Titanic, along with the information whether they survived the iceberg collision or not.

```python
# Imports
import numpy as np
import pandas as pd

# Reading csv data
titanic = pd.read_csv('train.csv')
titanic.shape
```

![](vari.png)

Above picture is from Kaggle

```python
# Removing unnecessary columns
titanic = titanic.drop(columns = ['Name', 'PassengerId', 'Ticket'])
titanic.shape
```

```python
titanic
```

```python
# Taking Cabin's first letter and making it the Deck. 
# Those without a letter shall be NaN. Then removing Cabin because it is an unnecessary column.
titanic["Deck"] = titanic["Cabin"].str[0]
titanic = titanic.drop("Cabin", axis = 1)
titanic.head(5)
```

```python
# Now we transform categorical values to numerical values
categorical_values = ["Sex", "Embarked", "Deck"]

for val in categorical_values:
    titanic[val] = titanic[val].astype('category')

cat_columns = titanic.select_dtypes(['category']).columns
titanic[cat_columns] = titanic[cat_columns].apply(lambda x: x.cat.codes)

titanic.replace(-1, np.NaN, inplace=True)

titanic.head(5)
```

## Imputation
Machine learning algorithm have problems with missing values so I'll fill them with means and median of the variable . This is not the best way to fill missing values, but if we take all the N/A values out of data we risk losing useful information. Of course there are better ways to fill missing values like making a predictive model.

```python
# Continuous variables imputation
for val in ["Age", "Fare", "Pclass", "SibSp"]:
    titanic[val].fillna(titanic[val].mean(), inplace=True)

# Categorical variables imputation
for val in categorical_values:
    titanic[val].fillna(titanic[val].median(), inplace=True)

# Printing means and medians
for val in ["Age", "Fare", "Pclass", "SibSp"]:
    print(val,": ",titanic[val].mean())
for val in categorical_values:
    print(val,": ",titanic[val].median()) 

titanic.head(5)
```

```python
titanic.to_csv('titanic.csv')
titanic.to_json('titanic.json', orient = 'table')
```

We should be very careful when choosing to impute missing data. Sometimes it can make big changes to the outcome of the analysis.

```python
import statistics as st
# medians of continous variables and modes of categorical variables to make an average human on board.
mSurvived = st.mode(titanic['Survived'])
mPclass = st.mode(titanic['Pclass'])
mSex = st.mode(titanic['Sex'])
mAge = np.mean(titanic['Age'])
mSibSp = np.mean(titanic['SibSp'])
mParch = np.mean(titanic['Parch'])
mFare = np.mean(titanic['Fare'])
mEmbarked = st.mode(titanic['Embarked'])
mDeck = st.mode(titanic['Deck'])

average = [mSurvived, mPclass, mSex, mAge, mSibSp, mParch, mFare,
       mEmbarked, mDeck]
ave = pd.DataFrame(['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked', 'Deck'], [average])
print("Average person on board: ", ave)
# 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked', 'Deck'
```

```python
# Now we visualize categorical variables.
def visualize_cat(x):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel(x)
    
    ax.set_ylabel('Passenger Count')

    tb = pd.pivot_table(data=titanic, values='Age', index=x, columns='Survived', aggfunc='count')
    bar1=tb[0]
    bar2=tb[1]
    
    i = np.arange(len(bar1))
    firstb = plt.bar(i, bar1, 0.5, color='b', label = 'Non-survivors')
    secondb = plt.bar(i, bar2, 0.5, color='y', bottom=bar1, label = 'Survivors')

    plt.title('Passenger count per ' + x)
    plt.legend(loc = 'upper center')
    plt.show()

# 0 (left) is male and 1 (right) is female
visualize_cat('Sex')
visualize_cat('Deck')
visualize_cat('Pclass')
visualize_cat('Embarked')
visualize_cat('SibSp')
visualize_cat('Parch')
```

## Supervised Machine Learning models and their accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
```

```python
titanic.head()
```

```python
# 80/20 train test split
# titanic.shape, 891/100*80,891/100*20
```

```python
# Prepare data for models.
# Splitting the Titanic data 80/20 train test
x_train = titanic.drop(labels = "Survived", axis = 1)[0:712]
y_train = titanic["Survived"][0:712]
x_test = titanic.drop("Survived", axis = 1)[713:890]
```

### Logistic regression
Logistic regression is also known as maximum-entropy classification (MaxEnt), logit regression or the log-linear classifier is a linear model for classification using logistic function. Commonly used for binary classification. In this case survivor and non survivor.

```python
#Logistic regression
logr = LogisticRegression( max_iter=1000)
logr.fit(x_train, y_train)
y_pred = logr.predict(x_test)
accuracy_logr = round(logr.score(x_train, y_train) * 100, 2)
accuracy_logr
```

### Support Vector Machine
Support vector machines are set of machine learning algorithms that can classify or regression analyze data. In Titanic data the algorithm draws a maximum-margin hyperplane that we can classify survivors and non-survivors. Below I used three kinds of kernels which means different kind of lines are drew. For instance linear kernel only draws straight lines and rbf kernel uses curvy lines according its algorithm. 

```python
# SVM linear kernel
svc = SVC(kernel = "linear")
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
accuracy_svc1 = round(svc.score(x_train, y_train)*100,2)
accuracy_svc1
#The best out of three kernels
```

```python
# SVM rbf kernel
svc = SVC(kernel = "rbf")
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
accuracy_svc2 = round(svc.score(x_train, y_train)*100,2)
accuracy_svc2
```

```python
# SVM polynomial kernel 
svc = SVC(kernel = "poly")
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
accuracy_svc3 = round(svc.score(x_train, y_train)*100,2)
accuracy_svc3
```

### k-Nearest Neighbor (k-NN)
In k-NN the point is classified according to its nearest neighbor. For instance if k = 1 then point is classified as the nearest neighbor point and with larger k it classifies the point according to majority of the k number of nearest neighbors.
This is widely used in unsupervised learning and pattern recognition.

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_knn = round(knn.score(x_train, y_train)*100,2)
accuracy_knn
```

### Naive Bayes Classifier
Naive bayes is probabilistic supervised learning classifier that uses Bayes theorem with strong independence assumptions between features (x_train or x_test). Below we use Gaussian Naive Bayes which uses normal distribution probability density function in the Bayes Theorem.

```python
gaus = GaussianNB()
gaus.fit(x_train, y_train)
y_pred = gaus.predict(x_test)
accuracy_gaus = round(gaus.score(x_train, y_train)*100,2)
accuracy_gaus
```

### Stochastic Gradient Descent
Stochastic gradient descent (SGD) is stricly speaking a optimization method. In another words a way to train a model under convex loss function. The model we are training in SGDClassifier is SVM.

```python
sgd = SGDClassifier()
sgd.fit(x_train, y_train)
y_pred = sgd.predict(x_test)
accuracy_sgd = round(sgd.score(x_train, y_train)*100,2)
accuracy_sgd
```

### Decision Tree
Decision treee is like a "flowchart" where every node splits into branches and also every node has a condition or a probability that splits the samples. At the end of the "flowchart" the sample is classified in this our case Survived variable. Below we try three different kind of decision trees classifiers. 

```python
decision_tree = DecisionTreeClassifier(criterion="gini")
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_test)
accuracy_decision_tree = round(decision_tree.score(x_train, y_train)*100,2)
accuracy_decision_tree
# We see that 98.46 is too accurate and I think it is overfitting. So I'll drop this from the comparison.
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(x_train, y_train)
y_pred = random_forest.predict(x_test)
accuracy_random_forest = round(random_forest.score(x_train, y_train)*100,2)
accuracy_random_forest
# We see that 98.46 is too accurate and I think it is overfitting. So I'll drop this from the comparison.
```

```python
decision_tree_regressor = DecisionTreeRegressor(criterion="poisson")
decision_tree_regressor.fit(x_train, y_train)
y_pred = decision_tree_regressor.predict(x_test)
accuracy_decision_tree_regressor = round(decision_tree_regressor.score(x_train, y_train)*100,2)
accuracy_decision_tree_regressor
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'k-Nearest Neighbor', 'Logistic Regression', 'Naive Bayes', 
              'Stochastic Gradient Decent', 
              'Decision Tree Regressor'],
    'Score': [accuracy_svc1, accuracy_knn,accuracy_logr,accuracy_gaus
              ,accuracy_sgd,accuracy_decision_tree_regressor]})
models.sort_values(by = "Score", ascending=False)
```

## Conclusion
In conclusion the best accuracy is from logistic regression.
